### ДЗ-2, задание 5

```python
# Клиент хочет дом с видом на набережную, как минимум с тремя ванными и с подвалом. 
# Сколько вариантов есть у клиента? Выберите дом клиенту
df[(df['waterfront'] == 1) & (df['bathrooms'] >= 3) & (df['sqft_basement'] != 0)].shape[0]
# Чтобы узнать, сколько есть строк и столбцов можно вызвать атрибут shape
```

```python
# в эталоне (df['bathrooms'] > 3), но это не включает минимум 3 ванны
df[(df['waterfront'] == 1) & (df['bathrooms'] > 3) & (df['sqft_basement'] != 0)].shape[0]
```

```python
# Клиент хочет дом либо с очень красивым видом из окна, либо с видом на набережную, в очень хорошем состоянии и год постройки не меньше 1980 года. В какой ценовом диапазоне будут дома?
cond_house = ((df['view'] == 5) | (df['waterfront'] == 1)) & (df['condition'] == 5) & (df['yr_built'] >= 1980)
client_choice = df[cond_house]
#df['yr_built'].min(), df['yr_built'].max()
client_choice['price'].min(), client_choice['price'].max()
```

```python
# Клиент хочет дом без подвала, с двумя этажами, стоимостью до 150000. Какая оценка по состоянию у таких домов в среднем?
client_want = df[(df['sqft_basement'] == 0) & (df['floors'] == 2) & (df['price'] < 150000)]['condition'].mean()
np.round(client_want, 3)
```
