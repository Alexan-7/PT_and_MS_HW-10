Провести дисперсионный анализ для определения того, есть ли различия среднего роста среди взрослых футболистов, хоккеистов и штангистов.
Даны значения роста в трех группах случайно выбранных спортсменов:
Футболисты: 173, 175, 180, 178, 177, 185, 183, 182.
Хоккеисты: 177, 179, 180, 188, 177, 172, 171, 184, 180.
Штангисты: 172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
y1 = np.array([173, 175, 180, 178, 177, 185, 183, 182])
y2 = np.array([177, 179, 180, 188, 177, 172, 171, 184, 180])
y3 = np.array([172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170])
y1, y2, y3
```

H0 - нулевая гипотеза - различий значений роста в группах нет.
H1 - алтернативная гипотеза - есть статистически значимые различия в значениях роста.

```python
# воспользуемся готовой функцией для выполнения однофакторного дисперсионного анализа
# В работе пользуйтесь функцией, не надо считать вручную - слова из лекции
f = stats.f_oneway(y1, y2, y3)
f
```

Теперь, чтобы сделать вывод, о том, оказывает ли вид спорта эффект на рост спортсмена, найдем по таблице Фишера https://univer-nn.ru/econometrica/Fisher-Student-table.pdf табличное значение критерия F

k - 1 = 3 - 1 = 2 и n - k = 28 - 3 = 25 - степени свободы; 
k − число групп, n – общее число значений роста.

Табличное значение критерия Фишера равно 3,38. Отметим его на схематично изображенном F-распределении Фишера.

Расчетное значение 5.5 попадает в область принятия альтернативной гипотезы H1. Т.е. мы делаем вывод, что вид спорта оказывает статистически значимый эффект на разницу в росте спортсменов.
